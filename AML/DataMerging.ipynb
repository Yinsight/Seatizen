```python
# Retrieve station data and train data
import json
with open('Stations.json') as f1:
    station_data = json.load(f1)

station_data = station_data['GetStationsJsonResult']

with open('trains.json') as f2:
    train_data = json.load(f2)

#print(station_data)
#print(train_data)
```

```python
import pandas as pd
df_station = pd.DataFrame(station_data)
df_station = df_station.drop(columns=['ZoneStationID'])
df_station.head(10)
```

```python
df_train = pd.DataFrame(train_data)
keep = ['BRANCH_ID','TRAIN_ID']
df_train = df_train[keep]
df_train = df_train.rename(columns={'BRANCH_ID': "BranchID", "TRAIN_ID": "TrainID"})
df_train.head(10)
```

```python
# Merge based on BranchID
df_merged = pd.merge(df_train, df_station, on=['BranchID'])
df_merged = df_merged.sample(frac=1)
df_merged = pd.concat([df_merged]*6, ignore_index=True).head(10000)
df_merged.head(10)
```

```python
import numpy as np
column_names = ['CarID', 'Month', 'Day', 'Hour', 'Passenger_Count']
# random numbers
mockup_data = pd.DataFrame(np.random.randint(0, 100, size=(10000, 5)), columns = column_names)
mockup_data['CarID'] = np.random.randint(1, 8, mockup_data.shape[0])
mockup_data['Month'] = np.random.randint(1, 12, mockup_data.shape[0])
mockup_data['Day'] = np.random.randint(1, 7, mockup_data.shape[0])
mockup_data['Hour'] = np.random.randint(0, 23, mockup_data.shape[0])
mockup_data['Passenger_Count'] = np.random.randint(1, 50, mockup_data.shape[0])
# populate columns with random numbers within a range
mockup_data.head(10)
```

```python
# Merge together to get the final dataset
data = pd.concat([df_merged.reset_index(drop=True), mockup_data.reset_index(drop=True)], axis=1)
data.head(10)
```

```python
# Drop station name since station id is already included
data = data.drop(columns=['StationName'])
data.to_csv('merged_data.csv')
```
